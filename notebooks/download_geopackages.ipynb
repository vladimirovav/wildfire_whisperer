# Download Geopackages

This notebook downloads and extracts the required geopackages for the project.

```python
import os
import time
import urllib.request
import zipfile
import shutil
import subprocess
import requests
from tqdm import tqdm

print('Starting geopackage download process...')

# Create shapefiles directory if it doesn't exist
os.makedirs('../data/shapefiles', exist_ok=True)
print('Shapefiles directory is ready')

# Function to download and extract a zip file with progress tracking
def download_and_extract(url, zip_filename, extract_dir, check_file=None):
    """Download and extract a zip file with progress tracking"""
    # Check if extraction directory exists and contains the check file
    if check_file and os.path.exists(os.path.join(extract_dir, check_file)):
        print(f'{extract_dir} already exists with {check_file}. Skipping download.')
        return True
    
    # Create extraction directory
    os.makedirs(extract_dir, exist_ok=True)
    
    # Download with progress tracking
    print(f"Downloading {url}...")
    
    # Download with requests and progress tracking
    try:
        # Get file size for progress bar
        response = requests.get(url, stream=True)
        total_size = int(response.headers.get('content-length', 0))
        
        # Download with progress bar
        with open(zip_filename, 'wb') as f:
            with tqdm(total=total_size, unit='B', unit_scale=True, desc="Downloading") as pbar:
                for data in response.iter_content(chunk_size=8192):
                    f.write(data)
                    pbar.update(len(data))
        
        print(f'Downloaded {zip_filename} successfully')
        # Extract the zip file
        with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
            zip_ref.extractall(extract_dir)
        print(f'Extraction to {extract_dir} complete.')
        # Remove the zip file
        os.remove(zip_filename)
        print(f'{zip_filename} removed.')
        return True
    except Exception as e:
        print(f'Error downloading {url}: {e}')
        return False

print('Setup complete!')
```

```python
# Download Natural Earth shapefiles
url = "https://naciscdn.org/naturalearth/10m/cultural/ne_10m_admin_1_states_provinces.zip"
zip_filename = "../data/shapefiles/natural_earth.zip"
extract_dir = "../data/shapefiles/natural_earth"
check_file = "ne_10m_admin_1_states_provinces.shp"

success = download_and_extract(url, zip_filename, extract_dir, check_file)
print("Natural Earth shapefiles download complete.")
```

```python
# Download Bioclimatic zones
url = "https://diffusion.mffp.gouv.qc.ca/Diffusion/DonneeGratuite/Foret/DONNEES_FOR_ECO_PROV/Classification_ecologique/02-Donnees/PROV/CLASSI_ECO_QC_GPKG.zip"
zip_filename = "../data/shapefiles/bioclimatic_zones.zip"
extract_dir = "../data/shapefiles/bioclimatic_zones"
check_file = "CLASSI_ECO_QC.gpkg"

success = download_and_extract(url, zip_filename, extract_dir, check_file)
print("Bioclimatic zones download complete.")
```
